```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
data_2017 = pd.read_csv("Data/meteorology2017.csv")
data_2018 = pd.read_csv("Data/meteorology2018.csv")
data_2019 = pd.read_csv("Data/meteorology2019.csv")
data_2020 = pd.read_csv("Data/meteorology2020.csv")

data = data_2020.append(data_2019.append(data_2018.append(data_2017)))
data["dateUTCShiftedDown"] = pd.to_datetime(data["dateUTCShiftedDown"])
data = data[data["id_station_id"] == "FAC"][["dateUTCShiftedDown","original","id_parameter_id"]]
```

```python
primavera = data[(data["dateUTCShiftedDown"].dt.day == 16) & ((data["dateUTCShiftedDown"].dt.month == 3) | 
                     (data["dateUTCShiftedDown"].dt.month == 4) | (data["dateUTCShiftedDown"].dt.month == 5))]

verano = data[(data["dateUTCShiftedDown"].dt.day == 16) & ((data["dateUTCShiftedDown"].dt.month == 6) | 
                     (data["dateUTCShiftedDown"].dt.month == 7) | (data["dateUTCShiftedDown"].dt.month == 8))]

otono = data[(data["dateUTCShiftedDown"].dt.day == 16) & ((data["dateUTCShiftedDown"].dt.month == 9) | 
                     (data["dateUTCShiftedDown"].dt.month == 10) | (data["dateUTCShiftedDown"].dt.month == 11))]

invierno = data[(data["dateUTCShiftedDown"].dt.day == 16) & ((data["dateUTCShiftedDown"].dt.month == 12) | 
                     (data["dateUTCShiftedDown"].dt.month == 1) | (data["dateUTCShiftedDown"].dt.month == 2))]
```

```python
para1 = primavera[primavera["id_parameter_id"] == "WSP"][["dateUTCShiftedDown","original"]]
para2 = verano[verano["id_parameter_id"] == "WSP"][["dateUTCShiftedDown","original"]]
para3 = otono[otono["id_parameter_id"] == "WSP"][["dateUTCShiftedDown","original"]]
para4 = invierno[invierno["id_parameter_id"] == "WSP"][["dateUTCShiftedDown","original"]]

dt = []
for year in [2017,2018,2019,2020]:
    for season in [para1,para2,para3,para4]:
        for hour in range(0,24):
            dt.append(season[(season["dateUTCShiftedDown"].dt.year == year) & 
                      (season["dateUTCShiftedDown"].dt.hour == hour)]["original"].mean())
fig, axs = plt.subplots(2, 2)
fig.set_size_inches(18.5,10,5)
seas = ["primavera","verano","otoño","invierno"]
for i in range(4):
    axs[0,0].plot(np.arange(0,24,1),dt[i*24:(i+1)*24],label=seas[i])
axs[0,0].set_title("WSP promediado por temporada en el año 2017")
for i in range(4,8):
    axs[0,1].plot(np.arange(0,24,1),dt[i*24:(i+1)*24],label=seas[i-4])
axs[0,1].set_title("WSP promediado por temporada en el año 2018")
for i in range(8,12):
    axs[1,0].plot(np.arange(0,24,1),dt[i*24:(i+1)*24],label=seas[i-8])
axs[1,0].set_title("WSP promediado por temporada en el año 2019")
for i in range(12,16):
    axs[1,1].plot(np.arange(0,24,1),dt[i*24:(i+1)*24],label=seas[i-12])
axs[1,1].set_title("WSP promediado por temporada en el año 2020")
axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
```
